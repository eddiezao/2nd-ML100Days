2. 資料從何而來 (tips：哪些資料可能會對你想問的問題產生影響 & 資料如何蒐集)
   資料如何蒐集:蒐集司機每天載客的地點,載客車密集度高之區域,發問卷給載客司機
   問題產生影響:可能有些冷門地點突然舉辦活動 造成當日該區載客數量變多

3. 蒐集而來的資料型態為何
   結構化資料型態為司機載客地點的文字敘述

4. 你要回答的問題，其如何評估 (tips：你的假設如何驗證)
   利用問卷調查或是司機載客地點等做回歸分析 敏感度分析  

```python
import numpy as np
import matplotlib.pyplot as plt


a=3
b=0.5

x = np.linspace(0,90,91)
y = a*(x+np.random.randn(91)*5)+b

plt.plot(x,y,'b.',label = 'data points')
plt.title("our data points")
plt.legend(loc = 2)
plt.show()

y_hat = a*x+b
plt.plot(x,y,'b.',label='data points')
plt.plot(x,y_hat,'r-',label ='prediction')
plt.title("our data points and prediction ")
plt.show()

def mean_square_error(y,y_hat):
    MSE = sum((y-y_hat)**2)/len(y)
    return MSE
MSE = mean_square_error(y,y_hat)
print("mean_square_error : %.3f "%(MSE))
```
